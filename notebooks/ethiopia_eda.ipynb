```python
import matplotlib.pyplot as plt

monthly_temp = df.groupby("DATE")["T2M"].mean()

plt.figure(figsize=(14, 6))
plt.plot(monthly_temp.index, monthly_temp.values)
plt.title("Daily Mean Temperature Trend in Ethiopia (2015–2026)")
plt.xlabel("Date")
plt.ylabel("Mean Temperature at 2m (°C)")
plt.grid(True)
plt.show()
```

```python
monthly_avg = df.groupby("MONTH")["T2M"].mean()

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(monthly_avg.index, monthly_avg.values, marker='o')

plt.title("Average Monthly Temperature Pattern in Ethiopia")
plt.xlabel("Month")
plt.ylabel("Temperature (°C)")
plt.xticks(range(1,13))
plt.grid(True)

plt.show()
```

## Temperature Trend Analysis

The monthly average temperature shows a clear seasonal pattern.  
Temperatures tend to peak during the warmer months and decline during cooler periods.  

This pattern reflects Ethiopia’s climatic seasonality and is important for understanding long-term temperature behavior and agricultural planning.

```python
monthly_rain = df.groupby("MONTH")["PRECTOTCORR"].sum()

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.bar(monthly_rain.index, monthly_rain.values)

plt.title("Monthly Total Rainfall in Ethiopia")
plt.xlabel("Month")
plt.ylabel("Total Rainfall (mm)")
plt.xticks(range(1,13))
plt.grid(axis='y')

plt.show()
```

## Rainfall Pattern Analysis

The monthly rainfall distribution shows a clear seasonal pattern.  
Rainfall increases significantly during the mid-year months, indicating the main rainy season.  

The peak rainfall period typically occurs between June and September, which aligns with Ethiopia’s Kiremt rainy season.  

This seasonal rainfall pattern is crucial for agriculture, water resource management, and climate planning.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select numeric columns
numeric_df = df.select_dtypes(include=["float64", "int64"])

# Correlation matrix
corr = numeric_df.corr()

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")

plt.title("Correlation Matrix of Climate Variables")
plt.show()
```

## Correlation Analysis

The correlation matrix reveals relationships between climate variables.

Strong positive correlations were observed between:
- (example: T2M and T2M_MAX)

Strong negative correlations were observed between:
- (example: T2M and RH2M)

These relationships highlight how temperature, humidity, and other variables interact in Ethiopia’s climate system.

Understanding these interactions is important for predicting weather patterns and assessing climate variability.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(df["PRECTOTCORR"].dropna(), bins=50)

plt.title("Distribution of Daily Rainfall")
plt.xlabel("Rainfall (mm)")
plt.ylabel("Frequency")

plt.show()
```

```python
import numpy as np

plt.figure(figsize=(8,5))
plt.hist(df["PRECTOTCORR"].dropna(), bins=50)

plt.yscale("log")

plt.title("Distribution of Daily Rainfall (Log Scale)")
plt.xlabel("Rainfall (mm)")
plt.ylabel("Frequency (log scale)")

plt.show()
```

## Rainfall Distribution Analysis

The distribution of daily rainfall is highly skewed.  
Most days record little to no rainfall, while a few days experience very high precipitation levels.  

This pattern is typical of tropical climates, where rainfall is concentrated in specific seasons rather than evenly distributed throughout the year.
